```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir frequência de amostragem

fs = 15.0
```

```python
T = 5.0         # tempo de simulação
n = int(T * fs) # número de amostras
tp = np.linspace(0, T, n, endpoint=False)
```

```python
tau = 1
xp = np.zeros(n)
largura = int(fs*tau)
xp[0:largura] = np.ones(largura)
x = np.concatenate((np.flipud(xp[1::]),xp),axis=None)

delta = 0
x = np.roll(x,delta)
```

```python
t = np.concatenate((-np.flipud(tp[1::]),tp),axis=None)
plt.plot(t,x)
```

```python
Xf = np.fft.fftshift(np.fft.fft(x))
print(np.max(Xf))
Xf = Xf/np.max(Xf)
f = np.linspace(-fs/2,fs/2,2*n-1)
```

```python
plt.plot(f,np.abs(Xf))
plt.grid()
plt.show()
```

```python
plt.plot(f,20*np.log10(np.abs(Xf)))
plt.grid()
plt.show()
```

```python
plt.plot(f,np.angle(Xf))
plt.grid()
plt.show()


```
