```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Definir frequência de amostragem


fs = 15.0
T = 5.0         # tempo de simulação
n = int(T * fs*2 ) # número de amostras
tp = np.linspace(-5, T, n, endpoint=False)

def yt_1(t):
    signal = np.zeros_like(tp)
    aux = 0
    for i in t:
        if 0 <= i < 2:
            signal[aux]= np.exp(-i)
        else:
            signal[aux]=0
        aux = aux + 1

    return signal
```

```python
yt_1(tp)
x = yt_1(tp)

plt.plot(tp,x)
plt.grid()
plt.show()
```

```python
#fft
Xf = np.fft.fftshift(np.fft.fft(x))
print(np.max(Xf))
print(Xf)
f = np.linspace(-fs/2,fs/2,n)
```

```python
#modulo
plt.plot(f,np.abs(Xf))
plt.grid()
plt.show()
```

```python
#modulo base log
plt.plot(f,20*np.log10(np.abs(Xf)))
plt.grid()
plt.show()
```

```python
#fase
plt.plot(f,np.angle(Xf))
plt.grid()
plt.show()
```

```python
iXf = np.fft.ifft(Xf)
print(np.max(iXf))
iXf = iXf/np.max(iXf)
```

```python
plt.plot(f,np.abs(iXf))
plt.grid()
plt.show()
```
